## 4. Clean the Dataset

Cleaning steps used:

- Remove duplicate rows.
- Fill missing `discount` values with `0`.
- Fill missing text values with `Not Available`.
- Replace seller values that contain black block/redacted text with `Not Available`.
- Drop the `videos` column because most rows are missing that value.
- Convert `final_price` into a numeric `price` column.
- Add a `quantity` column because the dataset does not already have one.
- Create `total_amount = price * quantity`.

```python
# Make a copy so the original DataFrame stays unchanged
df_clean = df.copy()

# Count duplicate rows before removing them
duplicate_count = df_clean.duplicated().sum()
print("Duplicate rows before cleaning:", duplicate_count)

# Remove duplicate rows
df_clean = df_clean.drop_duplicates()

print("Rows after removing duplicates:", df_clean.shape[0])
```

```python
# Fill missing discount values with 0
df_clean["discount"] = df_clean["discount"].fillna(0)

# Fill missing text values with a simple message
df_clean["what_customers_said"] = df_clean["what_customers_said"].fillna("Not Available")
df_clean["seller_name"] = df_clean["seller_name"].fillna("Not Available")
df_clean["seller_information"] = df_clean["seller_information"].fillna("Not Available")
df_clean["variations"] = df_clean["variations"].fillna("Not Available")

# Some seller names and seller addresses have black strips/redacted text.
# These values are not useful, so we replace them with Not Available.
black_block = chr(9608)

seller_name_hidden = df_clean["seller_name"].str.contains(black_block)
seller_info_hidden = df_clean["seller_information"].str.contains(black_block)

print("Seller names with black strips:", seller_name_hidden.sum())
print("Seller information with black strips:", seller_info_hidden.sum())

df_clean.loc[seller_name_hidden, "seller_name"] = "Not Available"
df_clean.loc[seller_info_hidden, "seller_information"] = "Not Available"

# Drop the videos column because many values are missing
df_clean = df_clean.drop("videos", axis=1)

print("Missing values and black strip values handled")
```

```python
# Convert final_price from text into a number
# Example: currency symbol + 3,995.00 becomes 3995.00
df_clean["price"] = df_clean["final_price"].astype(str)

# Remove quote marks and commas first
df_clean["price"] = df_clean["price"].str.replace('"', "", regex=False)
df_clean["price"] = df_clean["price"].str.replace(",", "", regex=False)

# Remove the first character because it is the currency symbol
df_clean["price"] = df_clean["price"].str[1:]

# Change the cleaned price text into numbers
df_clean["price"] = df_clean["price"].astype(float)

# The dataset does not have a quantity column, so we use 1 for each product
df_clean["quantity"] = 1

# Create the derived total_amount column
df_clean["total_amount"] = df_clean["price"] * df_clean["quantity"]

df_clean[["final_price", "price", "quantity", "total_amount"]].head()
```

## 4b. Clean and Expand Rating Details (amount_of_stars)

The amount_of_stars column has a dictionary string in it. I am using json_normalize to expand it out into separate columns.

```python
# The amount_of_stars column contains text in JSON format
# This is hard to analyze, so we convert it to separate numeric columns using json.loads
import json
stars_df = pd.json_normalize(df_clean["amount_of_stars"].apply(json.loads))

# Reset index to safely concatenate columns side-by-side
df_clean = df_clean.reset_index(drop=True)
df_clean = pd.concat([df_clean, stars_df], axis=1)

# Show a preview of the newly expanded star columns alongside the original column
df_clean[["amount_of_stars", "1_star", "2_stars", "3_stars", "4_stars", "5_stars"]].head()
```

## 4c. Clean and Expand Product Details (product_details)

The product_details column has json strings. I am splitting them out into separate columns for description, material_and_care, and size_and_fit.

```python
# Parse product_details JSON keys into separate text columns
details_df = pd.json_normalize(df_clean["product_details"].apply(json.loads))

# Fill any missing values in the expanded details with 'Not Available'
details_df = details_df.fillna("Not Available")

# Concatenate these new detail columns to our main DataFrame
df_clean = pd.concat([df_clean, details_df], axis=1)

# Show a preview of the original product_details and the newly extracted columns
df_clean[["product_details", "description", "material_and_care", "size_and_fit"]].head()
```

```python
# Check missing values again after cleaning
df_clean.isnull().sum()
```

## 5. Basic Data Operations

```python
# Filter rows where rating is 4.0 or higher
high_rated_products = df_clean[df_clean["rating"] >= 4.0]

high_rated_products.head()
```

```python
# Select only a few useful columns
selected_columns = ["product_id", "title", "category", "price", "quantity", "total_amount", "rating", "5_stars", "description"]

small_table = df_clean[selected_columns]
small_table.head()
```

## 6. Save the Cleaned Dataset

```python
# Save the cleaned data into a new CSV file
output_file = "ecommerce_cleaned.csv"
df_clean.to_csv(output_file, index=False)

print("Cleaned dataset saved as", output_file)
```
